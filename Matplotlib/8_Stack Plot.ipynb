# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,6,3,8]
plt.stackplot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,3,6,3,8]
y2=[2,4,1,5,1]
y3=[5,2,7,1,9]
l=["depart1","depart2","depart3"]
plt.stackplot(x,y1,y2,y3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,3,6,3,8]
y2=[2,4,1,5,1]
y3=[5,2,7,1,9]
l=["depart1","depart2","depart3"]
plt.stackplot(x,y1,y2,y3,labels=l,colors=['r','m','y'])
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,3,6,3,8]
y2=[2,4,1,5,1]
y3=[5,2,7,1,9]
l=["depart1","depart2","depart3"]
plt.stackplot(x,y1,y2,y3,labels=l,baseline="sym")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,3,6,3,8]
y2=[2,4,1,5,1]
y3=[5,2,7,1,9]
l=["depart1","depart2","depart3"]
plt.stackplot(x,y1,y2,y3,labels=l)
plt.title("Sales By Depart")
plt.legend()
plt.show()
```
