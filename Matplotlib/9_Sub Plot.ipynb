# Sub Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(x,y,color='m')

plt.subplot(2,2,2)
plt.pie([1],colors='y')

x1=[60,70,80,90,100]
plt.subplot(2,2,3)
plt.pie(x1)

a1=['a','b','c','d','e']
a2=[10,20,30,40,50]
plt.subplot(2,2,4)
plt.bar(a1,a2)
plt.show()
```
