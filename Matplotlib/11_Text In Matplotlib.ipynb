```python
import matplotlib.pyplot as plt
```

```python
a=[1,2,3,4,5]
b=[2,4,2,6,3]
plt.plot(a,b)
plt.title("Sales Report")
plt.xlabel("By Years")
plt.ylabel("By Product")
plt.text(3,4.5,"Average",bbox={"facecolor":'y'})
plt.annotate("Low",xy=(2.5,3),xytext=(4,4),arrowprops=dict(facecolor='r',shrink=100))
plt.legend(["Final Report"],loc=9,facecolor='m',edgecolor='y',shadow=True)
plt.show()
```
