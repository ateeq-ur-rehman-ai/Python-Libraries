# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":",markerfmt="r*")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":",bottom=10)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":",basefmt='g')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":",basefmt='g',label="Pyhton")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,3,1,4,6,2]
plt.stem(x,y,linefmt=":",label="Pyhton",basefmt='g',orientation="horizontal")
plt.legend()
plt.show()
```
