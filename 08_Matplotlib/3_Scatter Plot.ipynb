# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
plt.scatter(x,y)
plt.title("Scatter Plot",fontsize=25)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
plt.scatter(x,y)
plt.title("Scatter Plot",fontsize=25)
plt.xlabel("Sales",fontsize=25)
plt.ylabel("By Years",fontsize=25)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
c=['r','g','b','y','m','k','c','r']
plt.scatter(x,y,color=c)
plt.title("Scatter Plot",fontsize=25)
plt.xlabel("Sales",fontsize=25)
plt.ylabel("By Years",fontsize=25)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
c=['r','g','b','y','m','k','c','r']
s=[230,320,200,110,450,290,460,210]
plt.scatter(x,y,color=c,sizes=s,alpha=0.4)
plt.title("Scatter Plot",fontsize=25)
plt.xlabel("Sales",fontsize=25)
plt.ylabel("By Years",fontsize=25)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
c=['r','g','b','y','m','k','c','r']
s=[230,320,200,110,450,290,460,210]
plt.scatter(x,y,color=c,sizes=s,marker="*",edgecolor="g",linewidth=1)
plt.title("Scatter Plot",fontsize=25)
plt.xlabel("Sales",fontsize=25)
plt.ylabel("By Years",fontsize=25)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,1,5,7,5,6]
colors=[23,45,37,86,94,25,26,27]
s=[230,320,200,110,450,290,460,210]
plt.scatter(x,y,c=colors,sizes=s,cmap="plasma")
t=plt.colorbar()
t.set_label("ColorBar",fontsize=25)
plt.title("Scatter Plot",fontsize=25)
plt.xlabel("Sales",fontsize=25)
plt.ylabel("By Years",fontsize=25)
plt.show() 
```
