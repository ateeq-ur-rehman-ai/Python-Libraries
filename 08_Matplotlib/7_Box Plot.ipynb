# Box Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,notch=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,widths=0.9)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,labels=["DataSet"])
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,whis=2)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,sym="g+",boxprops=dict(color='r'),capprops=dict(color='m'))
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
y=[10,20,30,40,50,60,70]
z=[x,y]
plt.boxplot(z,showmeans=True,boxprops=dict(color='r'),capprops=dict(color='m'),whiskerprops=dict(color="y"))
plt.show()
```
