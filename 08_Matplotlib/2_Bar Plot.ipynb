```python
x=["Java","Html","Python","JavaScript","C"]
y=[78,23,89,70,65]
z=[37,92,68,80,43]
c=['c','r','b','y','g']
plt.xlabel("Languages",fontsize=20)
plt.ylabel("Use By Percentage",fontsize=20)
plt.title("Languages And Their Usage",fontsize=20)
plt.bar(x,y,color='b',label="BackEnd Languages.")
plt.bar(x,z,color='r',label="FrontEnd Languages.")
plt.legend()
plt.show
```

```python
import numpy as np 

x=["Java","Html","Python","JavaScript","C"]
y=[78,23,89,70,65]
z=[37,92,68,80,43]

width=0.4
a=np.arange(len(x))
a1=[j+width for j in a]

c=['c','r','b','y','g']

plt.xlabel("Languages",fontsize=20)
plt.ylabel("Use By Percentage",fontsize=20)
plt.title("Languages And Their Usage",fontsize=20)

plt.bar(a,y,width,color='b',label="BackEnd Languages.")
plt.bar(a1,z,width,color='r',label="FrontEnd Languages.")

plt.xticks(a+width/2,x,rotation=20)
plt.legend()
plt.show
```

```python
x = ["Java", "Html", "Python", "JavaScript", "C"]
y = [78, 23, 89, 70, 65]   # Backend
z = [37, 92, 68, 80, 43]   # Frontend

height = 0.4  

p = np.arange(len(x))
plt.xlabel("Use By Percentage", fontsize=20)
plt.ylabel("Languages", fontsize=20)
plt.title("Languages And Their Usage", fontsize=20)

plt.barh(p - height/2, y, height, color='b', label="BackEnd Languages")
plt.barh(p + height/2, z, height, color='r', label="FrontEnd Languages")

plt.yticks(p, x,rotation=20)
plt.legend()
plt.show()
```
