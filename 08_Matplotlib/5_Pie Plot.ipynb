# Pie Plot 

```python
import matplotlib.pyplot as plt 
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%')
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.3,labeldistance=1.2)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',startangle=90)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.3,startangle=90,textprops={"fontsize":15})
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.3,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":15})
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.3,rotatelabels=True)
plt.show()
```

```python
x=[10,20,30,40]
y=["Javascript","Java","SQL","Python"]
ex=[0.0,0.2,0.0,0.0]
c=['g','r','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=0.7)
plt.title("Language Usage By Percentage")
plt.legend(loc=2)
plt.show()
```

```python
x=[10,20,30,40]
x1=[50,60,70,80]
y=["Javascript","Java","SQL","Python"]
c=['b','g','m','y']
plt.pie(x,labels=y,radius=1.4)
plt.pie(x1,labels=y,radius=0.9,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
x1=[50,60,70,80]
y=["Javascript","Java","SQL","Python"]
c=['b','g','m','y']
plt.pie(x,labels=y,radius=1.4)
plt.pie([1],colors='w',radius=0.7)
plt.show()
```

```python
x=[10,20,30,40]
x1=[50,60,70,80]
y=["Javascript","Java","SQL","Python"]
c=['b','g','m','y']
plt.pie(x,labels=y,radius=1.4)
a=plt.Circle(xy=(0,0),radius=0.8,facecolor='w')
plt.gca().add_artist(a)
plt.show()
```
